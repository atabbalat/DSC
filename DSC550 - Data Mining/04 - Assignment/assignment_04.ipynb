```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sn
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error as mae, r2_score as r2, mean_squared_error as mse
from math import sqrt
from sklearn import linear_model
```

```python
# Load the data as a Pandas data frame and ensure that it imported correctly.
df = pd.read_csv('auto-mpg.csv')
df
```

```python
# Remove the car name column.
df = df.drop(columns=['car name'])
df
```

```python
# Column name adjustments
df = df.rename(columns={'model year' : 'modelyear'})
df
```

```python
# The horsepower column values likely imported as a string data type. Figure out why and replace any strings with the column mean.
## Filtering for ? in dataset
df_nanfilter = df[df['horsepower'] == '?']
df_nanfilter
```

```python
## replacing ? with Nan
df['horsepower'] = df['horsepower'].replace('?', np.nan)
df = df.astype({'horsepower' : 'float64'})
df.dtypes
```

```python
## calculating the mean and replacing NaNs with it
mean_hp = df['horsepower'].mean()
df['horsepower'] = df['horsepower'].fillna(mean_hp)
df
```

```python
# Create dummy variables for the origin column.
df['dummy1'] = df['origin'].apply(lambda x: 1 if x == 1 else 0)
df['dummy2'] = df['origin'].apply(lambda x: 1 if x == 2 else 0)
df['dummy3'] = df['origin'].apply(lambda x: 1 if x == 3 else 0)
df
```

```python
df.info()
```

```python
# Create a correlation coefficient matrix and/or visualization. Are there features highly correlated with mpg?
## Correlation
correlation = df.corr()
correlation
```

```python
## mpg correlation
correlation['mpg']
```

```python
sn.heatmap(correlation, annot=True)
sn.set(rc = {'figure.figsize' : (30, 20)})
plt.show()
```

We can tell that most of our variables are negatively correlated to mpg. For example:
* when horsepower goes up mpg goes down
* when weight of the car increases mpg will decrease
* The more cylinders we have or size of cylinder increases we will get less mpg

```python
# Plot mpg versus weight. Analyze this graph and explain how it relates to the corresponding correlation coefficient.
plt.scatter(df['mpg'], df['weight'])
plt.xlabel('MPG')
plt.ylabel('Weight')
plt.title('MPG vs Weight Plot')
plt.show()
```

Plot acknowledges correlation function applied in heatmap. When weight goes up, MPG go down.

```python
# Randomly split the data into 80% training data and 20% test data, where your target is mpg.
## Creating a dataframe that excludes mpg
df_exc_mpg = df[['cylinders', 'displacement', 'horsepower', 'weight',
       'acceleration', 'modelyear', 'origin', 'dummy1', 'dummy2', 'dummy3']]
df_exc_mpg
```

```python
## Splitting the data between 80% training and 20% testing
x_train, x_test, y_train, y_test = train_test_split(df_exc_mpg, df[['mpg']], test_size=.2)
print('Training data shape:')
print(x_train.shape)
print(y_train.shape)
print('\nTesting data shape:')
print(x_test.shape)
print(y_test.shape)
```

```python
# Train an ordinary linear regression on the training data.
model = LinearRegression() # creating LR model
model.fit(x_train, y_train) # Training model
```

```python
# Calculate R2, RMSE, and MAE on both the training and test sets and interpret your results.
## Training data calcs
y_predict_train = model.predict(x_train) # predicting training dataset
r2_train = r2(y_train, y_predict_train) # calculating R2
rmse_train = sqrt(mse(y_train, y_predict_train)) # calculating RMSE
mae_train = mae(y_train, y_predict_train) # calculating MAE

print('Training data R2, RMSE, MAE:')
print(f'R2: {r2_train}')
print(f'RMSE: {rmse_train}')
print(f'MAE: {mae_train}')
```

* R2 is at 82.0% meaning that our predicting is fairly close to observations
* RMSE at 3.36 meaning our predictions arent far from the observation values

```python
## Testing data calcs
y_predict_test = model.predict(x_test) # predicting training dataset
r2_test = r2(y_test, y_predict_test) # calculating R2
rmse_test = sqrt(mse(y_test, y_predict_test)) # calculating RMSE
mae_test = mae(y_test, y_predict_test) # calculating MAE

print('Testing data R2, RMSE, MAE:')
print(f'R2: {r2_test}')
print(f'RMSE: {rmse_test}')
print(f'MAE: {mae_test}')
```

* R2 is at 83.9% meaning that our predicting is fairly close to observations
* RMSE at 2.93 meaning our predictions arent far from the observation values

```python
# Pick another regression model and repeat the previous two steps. Note: Do NOT choose logistic regression as it is more like a classification model.
## Will use Ridge Regression
ridge_model = linear_model.Ridge().fit(x_train, y_train) # fitting Ridge model
```

```python
# Ridge -- Calculate R2, RMSE, and MAE on both the training and test sets and interpret your results.
## Ridge -- Training data calcs
r_y_predict_train = ridge_model.predict(x_train)
r_r2_train = r2(y_train, r_y_predict_train) # calculating R2
r_rmse_train = sqrt(mse(y_train, r_y_predict_train)) # calculating RMSE
r_mae_train = mae(y_train, r_y_predict_train) # calculating MAE

print('Ridge -- Training data R2, RMSE, MAE:')
print(f'R2: {r_r2_train}')
print(f'RMSE: {r_rmse_train}')
print(f'MAE: {r_mae_train}')
```

Almost an identical answer to the exercise above, which means the findings are the same for the training dataset.

```python
## Ridge -- Testing data calcs
r_y_predict_test = ridge_model.predict(x_test)
r_r2_test = r2(y_test, r_y_predict_test) # calculating R2
r_rmse_test = sqrt(mse(y_test, r_y_predict_test)) # calculating RMSE
r_mae_test = mae(y_test, r_y_predict_test) # calculating MAE

print('Ridge -- Test data R2, RMSE, MAE:')
print(f'R2: {r_r2_test}')
print(f'RMSE: {r_rmse_test}')
print(f'MAE: {r_mae_test}')
```

Almost an identical answer to the exercise above, which means the findings are the same for the testing dataset.
