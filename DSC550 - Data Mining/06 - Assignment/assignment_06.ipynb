#### Part 1: PCA and Variance Threshold in a Linear Regression

```python
# Import the housing data as a data frame and ensure that the data is loaded properly.
df = pd.read_csv('train.csv')
df
```

```python
# Drop the "Id" column and any features that are missing more than 40% of their values.
## Creating df to calculate missing NaNs over 40%
df_pct_col = pd.DataFrame({'Column' : df.columns, 'Percent' : df.isnull().sum() / len(df)})
df_40_col = df_pct_col[df_pct_col['Percent'] > 0.4]
df_40_col
```

```python
## Dropping columns shown in table above including ID
df = df.drop(['Id', 'Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature'] , axis=1)
df
```

```python
## NaN check on columns containing NaNs after drop
df_nan_col = pd.DataFrame({'Column' : df.columns, 'Percent' : df.isnull().sum() / len(df)})
df_bef_col = df_nan_col[df_nan_col['Percent'] > 0]
df_bef_col
```

```python
## Shape check for column amount after applying drop
print(f'Columns with NaNs: {df_bef_col.shape}')
```

```python
# For numerical columns, fill in any missing data with the median value.
df = df.fillna(df.median())
df
```

```python
## new percent NaN table check after applying median
df_pct_col_med = pd.DataFrame({'Column' : df.columns, 'Percent' : df.isnull().sum() / len(df)})
df_med_col = df_pct_col_med[df_pct_col_med['Percent'] > 0]
df_med_col
```

```python
print(f'Columns with NaNs after applying Median: {df_med_col.shape}')
```

```python
## Applying mode for the 11 columns left with NaNs
df['MasVnrType'] = df['MasVnrType'].fillna(df['MasVnrType'].mode()[0])
df['BsmtQual'] = df['BsmtQual'].fillna(df['BsmtQual'].mode()[0])
df['BsmtCond'] = df['BsmtCond'].fillna(df['BsmtCond'].mode()[0])
df['BsmtExposure'] = df['BsmtExposure'].fillna(df['BsmtExposure'].mode()[0])
df['BsmtFinType1'] = df['BsmtFinType1'].fillna(df['BsmtFinType1'].mode()[0])
df['BsmtFinType2'] = df['BsmtFinType2'].fillna(df['BsmtFinType2'].mode()[0])
df['Electrical'] = df['Electrical'].fillna(df['Electrical'].mode()[0])
df['GarageType'] = df['GarageType'].fillna(df['GarageType'].mode()[0])
df['GarageFinish'] = df['GarageFinish'].fillna(df['GarageFinish'].mode()[0])
df['GarageQual'] = df['GarageQual'].fillna(df['GarageQual'].mode()[0])
df['GarageCond'] = df['GarageCond'].fillna(df['GarageCond'].mode()[0])
df
```

```python
## new percent NaN table check after applying mode
df_pct_col_mod = pd.DataFrame({'Column' : df.columns, 'Percent' : df.isnull().sum() / len(df)})
df_mod_col = df_pct_col_mod[df_pct_col_mod['Percent'] > 0]
df_mod_col
```

```python
print(f'Columns with NaNs after applying Mode: {df_mod_col.shape}')
```

```python
# Convert the categorical columns to dummy variables.
df = pd.get_dummies(df)
df
```

```python
# Split the data into a training and test set, where the SalePrice column is the target.
x = df.drop(['SalePrice'], axis = 1)
y = df['SalePrice']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)

print('Split data shape:')
print(x_train.shape)
print(x_test.shape)
```

```python
# Run a linear regression and report the R2-value and RMSE on the test set.
model = LinearRegression() # model creation
model.fit(x_train, y_train) # model training
y_pred = model.predict(x_test) # model prediction
```

```python
# RMSE and R2
print(f'RMSE:{sqrt(mean_squared_error(y_test, y_pred))}')
print(f'R2: {r2_score(y_test, y_pred)}')
```

```python
# Fit and transform the training features with a PCA so that 90% of the variance is retained (see section 9.1 in the Machine Learning with Python Cookbook).
scaler = StandardScaler() # standard scaler created
features = scaler.fit_transform(x_train) # feature standardized
pca = PCA(n_components=0.9, whiten=True) # PCS created retaining 90%
features_pca = pca.fit_transform(features) # PCA applied
```

```python
# How many features are in the PCA-transformed matrix?
print(f'Features (Original): {features.shape}')
print(f'Features (PCA): {features_pca.shape}')
```

```python
# Transform but DO NOT fit the test features with the same PCA.
test_features = scaler.transform(x_test)
test_features_pca = pca.transform(test_features)
```

```python
# Repeat step 7 with your PCA transformed data.
model = LinearRegression() # model creation
model.fit(features_pca, y_train) # model training
y_pred_pca = model.predict(test_features_pca) # model prediction
```

```python
## RMSE and R2
print(f'RMSE:{sqrt(mean_squared_error(y_test, y_pred_pca))}')
print(f'R2: {r2_score(y_test, y_pred_pca)}')
```

```python
# Take your original training features (from step 6) and apply a min-max scaler to them.
scale = preprocessing.MinMaxScaler(feature_range=(0, 1))
feature_scale = scale.fit_transform(x_train)
```

```python
# Find the min-max scaled features in your training set that have a variance above 0.1 (see Section 10.1 in the Machine Learning with Python Cookbook).
thresh = VarianceThreshold(threshold=0.1)
feature_high = thresh.fit_transform(feature_scale)
```

```python
# Transform but DO NOT fit the test features with the same steps applied in steps 11 and 12.
feature_scale_test = scale.transform(x_test)
high_variance = thresh.transform(feature_scale_test)
```

```python
# Repeat step 7 with the high variance data.
model = LinearRegression() # model creation
model.fit(feature_high, y_train) # model training
y_pred_scale = model.predict(high_variance) # model prediction
```

```python
## RMSE and R2
print(f'RMSE:{sqrt(mean_squared_error(y_test, y_pred_scale))}')
print(f'R2: {r2_score(y_test, y_pred_scale)}')
```

```python
# Summarize your findings.
print(f'LR (No Transformation: )')
print(f'RMSE:{sqrt(mean_squared_error(y_test, y_pred))}')
print(f'R2: {r2_score(y_test, y_pred)}\n')
print(f'LR (PCS Transformation: )')
print(f'RMSE:{sqrt(mean_squared_error(y_test, y_pred_pca))}')
print(f'R2: {r2_score(y_test, y_pred_pca)}\n')
print(f'LR (MinMax Transformation: )')
print(f'RMSE:{sqrt(mean_squared_error(y_test, y_pred_scale))}')
print(f'R2: {r2_score(y_test, y_pred_scale)}')
```

#### Part 2: Categorical Feature Selection

```python
# Import the data as a data frame and ensure it is loaded correctly.
df_m = pd.read_csv('mushrooms.csv')
df_m
```

```python
# Convert the categorical features (all of them) to dummy variables.
df_m_dummy = pd.get_dummies(df_m, columns=df_m.drop(['class'], axis=1).columns)
df_m_dummy['class_val'] = df_m_dummy['class'].apply(lambda x: 1 if x == 'e' else 0)
df_m_dummy['class'] = df_m_dummy['class'].apply(lambda x: 'e' if x == 'e' else 'p')
df_m_dummy
```

```python
df_m_dummy.columns
```

```python
# Split the data into a training and test set.
xm = df_m_dummy.drop(['class', 'class_val'], axis = 1)
ym = df_m_dummy[['class_val']]
xm_train, xm_test, ym_train, ym_test = train_test_split(xm, ym, test_size = 0.2)

print('Training data shape:')
print(xm_train.shape)
print(ym_train.shape)

print('Testing data shape:')
print(xm_test.shape)
print(ym_test.shape)
```

```python
# Fit a decision tree classifier on the training set.
tree_model = DecisionTreeClassifier()
tree_model.fit(xm_train, ym_train)
ym_pred = tree_model.predict(xm_test)
```

```python
# Report the accuracy and create a confusion matrix for the model prediction on the test set.
print(f'Accuracy: {metrics.accuracy_score(ym_test, ym_pred)}\n')
print(f'Matrix: ')
print(pd.DataFrame(metrics.confusion_matrix(ym_test, ym_pred)).set_axis(['E', 'P'], axis=1).rename({0 : 'E', 1 : 'P'}, axis=0))
```

```python
# Create a visualization of the decision tree.
feature_names = list(df_m_dummy.drop(['class'], axis=1).columns)
target_names = [str(s) for s in df_m_dummy['class']]
tree = t.plot_tree(tree_model, filled=True, feature_names=feature_names, class_names=target_names)
```

```python
# Use a χ2-statistic selector to pick the five best features for this data (see section 10.4 of the Machine Learning with Python Cookbook).
best = SelectKBest(chi2, k=5)
features_kbest = best.fit_transform(xm_train, ym_train)
print(f'Features (Original): {xm_train.shape}')
print(f'Features (Reduced): {features_kbest.shape}')
```

```python
# Which five features were selected in step 7? Hint: Use the get_support function.
selected_features = best.get_support(indices=True).tolist()
print(f'Features Selected: \n{df_m_dummy.iloc[:,selected_features].columns}')
```

```python
# Repeat steps 4 and 5 with the five best features selected in step 7.
## Split the data into a training and test set.
x5 = df_m_dummy[['odor_c', 'odor_m', 'gill-size_n', 'stalk-surface-above-ring_f', 'stalk-surface-below-ring_f']]
y5 = df_m_dummy[['class_val']]
x5_train, x5_test, y5_train, y5_test = train_test_split(x5, y5, test_size = 0.2)

print('Training data shape:')
print(x5_train.shape)
print(y5_train.shape)

print('Testing data shape:')
print(x5_test.shape)
print(y5_test.shape)
```

```python
## Fit a decision tree classifier on the training set.
tree_model_5 = DecisionTreeClassifier()
tree_model_5.fit(x5_train, y5_train)
y5_pred = tree_model_5.predict(x5_test)
```

```python
## Report the accuracy and create a confusion matrix for the model prediction on the test set.
print(f'Accuracy: {metrics.accuracy_score(y5_test, y5_pred)}\n')
print(f'Matrix: ')
print(pd.DataFrame(metrics.confusion_matrix(y5_test, y5_pred)).set_axis(['E', 'P'], axis=1).rename({0 : 'E', 1 : 'P'}, axis=0))
```

```python
# Create a visualization of the decision tree.
feature_names_5 = list(df_m_dummy[['odor_c', 
                                   'odor_m', 
                                   'gill-size_n', 
                                    'stalk-surface-above-ring_f', 
                                    'stalk-surface-below-ring_f']].columns)
target_names_5 = [str(s) for s in df_m_dummy['class']]
tree = t.plot_tree(tree_model_5, filled=True, feature_names=feature_names_5, class_names=target_names_5)
```
