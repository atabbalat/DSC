```python
import pandas as pd
import numpy as np
```

```python
# Load the dataset as a Pandas data frame
df = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
df
```

```python
# Display the first ten rows of data.
df.head(10)
```

```python
# Find the dimensions (number of rows and columns) in the data frame. What do these two numbers represent in the context of the data?
print(df.shape)
print(f'We have {df.shape[0]} rows and {df.shape[1]} columns')
```

```python
# Find the top five games by critic score.
df_critic = df.sort_values(by=['Critic_Score'], ascending=False)
df_critic.head(5)
```

```python
# Find the number of video games in the data frame in each genre.
df_count = df.groupby('Genre').count()
df_count
```

```python
# Find the first five games in the data frame on the SNES platform.
df_snes = df.sort_values(by=['Year_of_Release'])
df_snes = df_snes[df_snes['Platform'] == 'SNES'].head(5)
df_snes
```

```python
# Find the five publishers with the highest total global sales. Note: You will need to calculate the total global sales for each publisher to do this.
df_sales = df.groupby('Publisher').sum().reset_index()
df_sales = df_sales[['Publisher', 'Global_Sales']]
df_sales = df_sales.sort_values(by=['Global_Sales'], ascending=False).head(5)
df_sales
```

```python
# Create a new column in the data frame that calculates the percentage of global sales from North America. Display the first five rows of the new data frame.
df_ratio = df
df_ratio['NA_Sales_Ratio'] = df_ratio['NA_Sales'] / df_ratio['Global_Sales']
df_ratio.head(5)
```

```python
# Find the number NaN entries (missing data values) in each column.
df.isna().sum()
```

```python
# Try to calculate the median user score of all the video games. You will likely run into an error because some of the user score entries are a non-numerical
# string that cannot be converted to a float. Find and replace this string with NaN and then calculate the median. Then, replace all NaN entries in the user
# score column with the median value.
df_median = df
print(f'User_Score type before update: {df_median["User_Score"].dtypes}')

df_median = df_median.replace('tbd', np.nan)
df_median = df_median.astype({'User_Score' : 'float64'})
print(f'User_Score type after update: {df_median["User_Score"].dtypes}')

user_median = df_median['User_Score'].median()
print(f'\nThe median for User_Score is: {user_median}')

df_median = df_median.replace(np.nan, user_median)
print(f'\nNan check for User_Score: {df_median["User_Score"].isna().sum()}')
df_median

```
