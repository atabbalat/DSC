##### Sources: Scaling

```python
# Scaling calculations
scaling_result = []

# Constants
hd = 1 / 10

# Helper function to calculate HD
def calculate_hd(data, hd_ratio):
    return math.ceil(data * 3 * hd_ratio)

# Daily Twitter Tweets (Uncompressed)
daily_twitter_tweets_un = (500000000 * 128 * 1) / 1000000000000
hd_daily_twitter_tweets_un = calculate_hd(daily_twitter_tweets_un, hd)
scaling_result.append(['Daily Twitter Tweets (Uncompressed)', daily_twitter_tweets_un, hd_daily_twitter_tweets_un])

# Daily Twitter Tweets (Snappy Compressed)
daily_twitter_tweets_snappy = daily_twitter_tweets_un / 1.5
hd_daily_twitter_tweets_snappy = calculate_hd(daily_twitter_tweets_snappy, hd)
scaling_result.append(['Daily Twitter Tweets (Snappy Compressed)', daily_twitter_tweets_snappy, hd_daily_twitter_tweets_snappy])

# Daily Instagram Photos
daily_instagram_photo = (100000000 * 0.75) * (0.00000315)
hd_daily_instagram_photo = calculate_hd(daily_instagram_photo, hd)
scaling_result.append(['Daily Instagram Photos', daily_instagram_photo, hd_daily_instagram_photo])

# Daily YouTube Videos
daily_youtube_video = (500 * 1440) * (878.9 / 1000000)
hd_daily_youtube_video = calculate_hd(daily_youtube_video, hd)
scaling_result.append(['Daily YouTube Videos', daily_youtube_video, hd_daily_youtube_video])

# Yearly calculations
days_per_year = 365

# Yearly Twitter Tweets (Uncompressed)
yearly_twitter_tweets_un = daily_twitter_tweets_un * days_per_year
hd_yearly_twitter_tweets_un = calculate_hd(yearly_twitter_tweets_un, hd)
scaling_result.append(['Yearly Twitter Tweets (Uncompressed)', yearly_twitter_tweets_un, hd_yearly_twitter_tweets_un])

# Yearly Twitter Tweets (Snappy Compressed)
yearly_twitter_tweets_snappy = daily_twitter_tweets_snappy * days_per_year
hd_yearly_twitter_tweets_snappy = calculate_hd(yearly_twitter_tweets_snappy, hd)
scaling_result.append(['Yearly Twitter Tweets (Snappy Compressed)', yearly_twitter_tweets_snappy, hd_yearly_twitter_tweets_snappy])

# Yearly Instagram Photos
yearly_instagram_photo = daily_instagram_photo * days_per_year
hd_yearly_instagram_photo = calculate_hd(yearly_instagram_photo, hd)
scaling_result.append(['Yearly Instagram Photos', yearly_instagram_photo, hd_yearly_instagram_photo])

# Yearly YouTube Videos
yearly_youtube_video = daily_youtube_video * days_per_year
hd_yearly_youtube_video = calculate_hd(yearly_youtube_video, hd)
scaling_result.append(['Yearly YouTube Videos', yearly_youtube_video, hd_yearly_youtube_video])

# Create DataFrame with scaling results
columns = ['Data Item', 'Size', '# HD']
df_scaling = pd.DataFrame(scaling_result, columns=columns)
df_scaling
```

##### Sources: Reliability

```python
# Source = https://www.backblaze.com/b2/hard-drive-test-data.html

df_reliability = df_scaling[['Data Item', '# HD']]
df_reliability = df_reliability[df_reliability['Data Item'].str.startswith('Yearly')]
df_reliability['Failures'] = df_reliability['# HD'].apply(lambda x: math.ceil(x * 0.0137))
df_reliability
```
